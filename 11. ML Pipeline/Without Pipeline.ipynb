```python
# Import Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Upload Dataset
df = pd.read_csv('Record.csv')
df
```

```python
# Remove the column which is not required
df.drop(columns = ['ID'],inplace = True)
df.head()
```

```python
# Train and split dataset
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Result']), df['Result'], test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
y_train
```

```python
y_train.shape
```

```python
# See if there is any missing data
df.isnull().sum()
```

```python
# Apply smple imputer to Age to compensate missing value
si_Age = SimpleImputer()
X_train_Age = si_Age.fit_transform(X_train[['Age']])
X_test_Age = si_Age.transform(X_test[['Age']])
X_train_Age
```

```python
X_test_Age
```

```python
# Gender is a nominal category so apply OneHot Encoder
ohe_Gender = OneHotEncoder (sparse_output = False,handle_unknown ='ignore')
X_train_Gender = ohe_Gender.fit_transform(X_train[['Gender']])
X_test_Gender = ohe_Gender.transform(X_test[['Gender']])
X_train_Gender
```

```python
X_test_Gender
```

```python
# Apply Ordinal Encoder to Grade since it is ordinal categorical value
oe_Grade = OrdinalEncoder(categories = [['High School','Undergraduate','Postgraduate']])
X_train_Grade = oe_Grade.fit_transform(X_train[['Grade']])
X_test_Grade = oe_Grade.fit_transform(X_test[['Grade']])
X_train_Grade
```

```python
X_test_Grade
```

```python
# Apply Label encoding for Result since it is output feature
le_Result = LabelEncoder()
y_train_Result = le_Result.fit_transform(y_train)
y_test_Result = le_Result.transform(y_test)
y_train_Result
```

```python
# Concatenate all values
X_train_new = np.concatenate(( X_train_Age, X_train_Gender, X_train_Grade), axis=1)
# also the test data
X_test_new = np.concatenate((X_test_Age, X_test_Gender, X_test_Grade), axis=1)

X_train_new.shape
```

```python
# Implementation of model
clf = DecisionTreeClassifier()
clf.fit(X_train_new, y_train)
```

```python
# Prediction
y_pred = clf.predict(X_test_new)
y_pred
```

```python
#accuracy
accuracy_score(y_pred, y_test)
```

```python
# Code for production
import pickle
pickle.dump(ohe_Gender,open('Machine Learning/ohe_Gender.pkl','wb'))
pickle.dump(oe_Grade,open('Machine Learning/oe_Grade.pkl','wb'))
pickle.dump(le_Result,open('Machine Learning/le_Result.pkl','wb'))
pickle.dump(clf,open('Machine Learning/clf.pkl','wb'))
```
