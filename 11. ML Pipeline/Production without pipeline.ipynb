```python
#Import Libraries
import pickle
import numpy as np
```

```python
# Loading file into the folder
ohe_Gender = pickle.load(open('Machine Learning/ohe_Gender.pkl','rb'))
oe_Grade = pickle.load(open('Machine Learning/oe_Grade.pkl','rb'))
le_Result = pickle.load(open('Machine Learning/le_Result.pkl','rb'))
clf = pickle.load(open('Machine Learning/clf.pkl','rb'))
```

```python
#Giving sample input to the test
test_input = np.array([18,'other','High School', 'Good'],dtype=object).reshape(1,4)
test_input
```

```python
# Apply ohe for test input
test_input_Gender = ohe_Gender.transform(test_input[:,1].reshape(1,1))
#Age
test_input_Age = test_input[:,0].reshape(1,1)
# Apply oe for grade
test_input_Grade = oe_Grade.transform(test_input[:,2].reshape(1,1))
# concatenate all 
test_input_new = np.concatenate((test_input_Gender, test_input_Age, test_input_Grade),axis=1)
```

```python
#Predict
clf.predict(test_input_new)
```
