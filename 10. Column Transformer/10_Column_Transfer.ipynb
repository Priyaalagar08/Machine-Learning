**Split data into train and test before doing encoders**

```python
from sklearn.model_selection import train_test_split
(X_train,X_test,y_train,y_test)= train_test_split(df.drop(columns=['Result']),df['Result'],test_size=0.2,random_state=2)
```

**Using all encoders seperately**

```python
#Simple Imputer for missing values
si = SimpleImputer()
X_train_Age = si.fit_transform(X_train[['Age']])
X_test_Age = si.transform(X_test[['Age']])
X_train_Age.shape
```

```python
# Ordinal Encoder for Ordinal categorical data
oe = OrdinalEncoder(categories=[['High School','Undergraduate','Postgraduate']])
X_train_Grade = oe.fit_transform(X_train[['Grade']])
X_test_Grade = oe.transform(X_test[['Grade']])
X_train_Grade.shape
```

```python
# OneHot Encoding for Nominal categorical data
ohe = OneHotEncoder(drop = 'first')
X_train_Gender = ohe.fit_transform(X_train[['Gender']])
X_test_Gender = ohe.transform(X_test[['Gender']])
X_train_Gender.shape
```

```python
#Extracting ID
X_train_ID = X_train[['ID']].values
X_test_ID = X_test[['ID']].values
X_train_ID.shape
```

```python
#Extracting Age
X_train_Age = X_train[['Age']].values
X_test_ID = X_test[['Age']].values
X_train_Age.shape
```

```python
X_train_transformed = np.concatenate((X_train_ID, X_train_Age, X_train_Gender.toarray(), X_train_Grade), axis=1)
# also the test data
X_test_transformed = np.concatenate((X_test_ID, X_test_Age, X_test_Gender.toarray(), X_test_Grade), axis=1)

X_train_transformed.shape
```

**Use of Column transfer**

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['Age']),
    ('tnf2',OrdinalEncoder(categories=[['High School','Undergraduate','Postgraduate']]),['Grade']),
    ('tnf3',OneHotEncoder(drop='first'),['Gender'])
],remainder='passthrough')
```
