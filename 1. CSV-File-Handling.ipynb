# Handling CSV File 

```python
import pandas as pd
```

```python
# Dataset uploaded from files
df = pd.read_csv("Machine Learning/restaurant_orders.csv")
df
```

```python
# Dataset uploaded from open url 
url = "https://people.sc.fsu.edu/~jburkardt/data/csv/hw_200.csv"
df = pd.read_csv(url)
df
```

```python
# Dataset uploaded from URL
df = pd.read_csv("Machine Learning/Medical cost dataset.txt", sep='\t')
df
```

```python
# To change the name of the headings
df = pd.read_csv("Machine Learning/Medical cost dataset.txt", sep='\t', names=['S.No','Age','Sex','BMI','Children','Smoker','Region','Charges'])
df
```

```python
#Index column
df = pd.read_csv("Machine Learning/Medical cost dataset.txt", sep='\t',index_col='age')
df
```

```python
# Use Column
df = pd.read_csv("Machine Learning/Medical cost dataset.txt", sep='\t',usecols=['age','bmi','charges'])
df
```

```python
# Skip rows
df = pd.read_csv("Machine Learning/Medical cost dataset.txt", sep='\t',skiprows=[0,1])
df
```

```python
# n rows
df = pd.read_csv("Machine Learning/Medical cost dataset.txt", sep='\t',nrows=50)
df
```

```python
# Handling dataset
df = pd.read_csv("Machine Learning/Medical cost dataset.txt", sep='\t').info()
df
```
